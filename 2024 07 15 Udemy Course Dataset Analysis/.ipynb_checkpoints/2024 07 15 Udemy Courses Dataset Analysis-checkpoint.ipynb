### 10. What are the maximum number of subscribers for each level of courses

```python
#First, let's just use value counts
data.head(1)
```

```python
data['level'].value_counts() #entries in total, not subscribers, but some initial indicator of data shape for confirmation
```

```python
data.groupby('level')['num_subscribers'].sum() #This is all subscribers across all courses
```

```python
data.groupby('level')['num_subscribers'].max() #max subscribers to a single course within each category
```
