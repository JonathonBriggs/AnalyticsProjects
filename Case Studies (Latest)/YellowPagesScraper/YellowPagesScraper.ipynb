```python
#Importing libraries needed
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
main_list = []

def extract(url):
    #url = 'https://www.yellowpages.com/search?search_terms=restaurant&geo_location_terms=OH&page=1'
    headers = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/128.0.0.0 Safari/537.36'}
    r = requests.get(url, headers=headers)
    soup = BeautifulSoup(r.content, 'html.parser')
    return soup.find_all('div', class_ = 'info')
#print(len(articles))

def transform(articles):
    for item in articles:
        try:
            name = item.find('a', class_ = 'business-name').text
        except:
            name = ''
        
        try:
            address = item.find('div', class_ = 'street-address').text
        except:
            address = ''

        try:
            locality = item.find('div', class_ = 'locality').text
        except:
            locality = ''
        
        try:
            phone = item.find('div', class_ = 'phones phone primary').text
        except:
            phone = ''
            
        business = {
            'name': name,
            'address': address,
            'locality': locality,
            'phone': phone
        }
        main_list.append(business)

def load():
    df = pd.DataFrame(main_list)
    df.to_csv('RestaurantList.csv', index=False)

for x in range(1,299):
    print(f'Getting Page {x}')
    articles = extract(f'https://www.yellowpages.com/search?search_terms=restaurant&geo_location_terms=OH&page={x}')
    transform(articles)
    time.sleep(5)

load()
print('Save to CSV complete')
```
